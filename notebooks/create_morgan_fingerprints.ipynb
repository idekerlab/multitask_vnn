```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
```

```python
drug_list = list(pd.read_csv("../data/training_files_av/drug2ind_av.txt", sep="\t", header=None, names=['I', 'D'])['D'])
```

```python
fp_size = 512

fp_df = pd.DataFrame(columns=range(fp_size))

for i, drug in enumerate(drug_list):
    try:
        mol = Chem.MolFromSmiles(drug)
        fingerprint = rdMolDescriptors.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=fp_size)
        fp_df.loc[i] = list(fingerprint)
    except:
        print(drug, mol)

fp_df.to_csv("../data/training_files_av/drug2fingerprint_av.txt", header=None, index=False)
```
